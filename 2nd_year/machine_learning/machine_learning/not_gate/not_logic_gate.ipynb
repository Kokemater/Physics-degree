# Gradient Descent 1
Not gate

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[0],[1]])
y = np.array([[1],[0]])
print(x)
print(y)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    sig = sigmoid(x)
    return sig * (1 - sig)

def MSE(y, a_2):
    return 0.5 * np.sum((a_2 - y)**2)
```

```python
# initializa params randomly
w_1 = np.random.rand()
w_2 = np.random.rand()
b_1 = np.random.rand()
b_2 = np.random.rand()

```

```python
Cs = []
iterations = [i for i in range(20000)]
learning_rate1 = 0.1
print(w_2)

for i in iterations:
    # forward propagation
    z_1 = w_1 * x + b_1
    a_1 = sigmoid(z_1)
    z_2 = w_2 * a_1 + b_2
    a_2 = sigmoid(z_2)
    # back propagation
    delta_2 =(a_2 - y) * sigmoid_derivative(z_2)
    delta_1 = w_2 * sigmoid_derivative(z_1)
    partial_w2 = delta_2 * a_1
    partial_b2 = delta_2 * 1
    partial_w1 = delta_1 * delta_2 * x
    partial_b1 = delta_1 * delta_2 * 1
    C = MSE(y, a_2)
    Cs.append(C)

    # gradient descent
    w_2 -= np.sum(partial_w2) * learning_rate1
    b_2 -= np.sum(partial_b2) * learning_rate1
    w_1 -= np.sum(partial_w1) * learning_rate1
    b_1 -= np.sum(partial_b1) * learning_rate1
```

```python
# 
plt.figure(figsize=(8, 6))
plt.plot(iterations, Cs, label='Sine wave', color='blue')
plt.title('MSE error')
plt.xlabel('n iterations')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
```
