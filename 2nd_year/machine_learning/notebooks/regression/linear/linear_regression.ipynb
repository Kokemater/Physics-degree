```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'x_y_values_noisy_with_errors.xlsx'  # Reemplaza con la ruta correcta
data = pd.read_excel(file_path)

# Mostrar los primeros registros del DataFrame
print(data.head())
```

```python
x = data.x
y = data.y
x_err = data.x_error
y_err = data.y_error
# Realizar la regresión lineal
coefficients = np.polyfit(x, y, 1)  # Grado 1 para una línea recta
slope, intercept = coefficients

print(f'Pendiente: {slope}')
print(f'Intersección: {intercept}')
```

```python

# Crear la línea de regresión
y_pred = slope * x + intercept

# Crear la gráfica
plt.figure(figsize=(8, 6))
plt.errorbar(x, y, xerr=x_err, yerr=y_err, fmt='o', color='blue', ecolor='gray', capsize=5, label='Datos con error')
plt.plot(x, y_pred, color='red', label=f'Regresión lineal: y = {slope:.2f}x + {intercept:.2f}')

# Etiquetas y título
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de datos y regresión lineal')
plt.legend()

# Mostrar la gráfica
plt.show()
```
