```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
```

```python

# Supongamos que 'df' es un DataFrame con las columnas 'Altura', 'Mean', y 'Error'
mat_data = scipy.io.loadmat('datos.mat')  # Reemplaza con la ruta a tu archivo .mat
Altura = np.array(mat_data["Altura"])
temperaturas = np.array(mat_data["T_medidas"])
print(temperaturas.shape)

np_mean = []
np_error = []
for i in range(Altura.size):
    np_mean.append(np.mean(temperaturas[:,i]))
    np_error.append(np.std(temperaturas[:,i]))

mean = np.array(np_mean)
error = np.array(np_error)
print(Altura.shape)
print(mean.shape)
print(error.shape)
```

```python
plt.figure(figsize=(8, 6))
plt.errorbar(Altura.T, mean.T, yerr=error, fmt='o', color='blue', ecolor='gray', capsize=5, label='Datos con error')

# Etiquetas y t√≠tulo
plt.xlabel('Altura')
plt.ylabel('Temperatura media')
plt.title('DATOS')
plt.legend()
```

```python

values = np.random.uniform(0, 100, (1, 5))
values = np.array([[280, 5.91, -12.18, 12.01, 2.521]])

x = Altura
y = mean
h_mean = (x * mean).sum() / mean.sum()
def f(values):
    y_pred = values[0,0] + values[0,1] * (x - h_mean) + values[0,2] * (x - h_mean)**2 + values[0,3] * (x - h_mean)**3 + values[0,4] * (x - h_mean)**4
    return y_pred
y_pred = f(values)
def mse(x, y, y_pred):
    return 1/x.size * np.sum((y - y_pred)**2)
error_mse = mse(x, y, y_pred)
error_mse
```

WoW ! Como se puede ver, estos valores aleatorios dan lugar a un error enorme, hagamos un descenso por gradiente para minimizarlo.

```python
def gradient(y_pred, y, x, h_mean):
    gradient = []
    for i in range(5):
        grad = (-2/x.size) * np.sum((y - y_pred) * (x - h_mean)**i)
        gradient.append(grad)
    return np.array(gradient)

```

```python

x = x.ravel()
y_pred = y_pred.ravel()
print(x.shape)
print(y.shape)
print(y_pred.shape)
```

```python
# Gradient Descent
learning_rate = 1e-11
mse_list = []
interations = [i for i in range(10000)]
for iteration in interations:
    mse_list.append(mse(x,y,y_pred))
    y_pred = f(values)
    grad = gradient(y_pred, y, x, h_mean)
    values -= learning_rate * grad

mse_list = np.array(mse_list)
print(f"MSE = {mse(x,y,y_pred)}")
```

```python
plt.figure(figsize=(8, 6))
plt.plot(interations, mse_list, color='red', label='Modelo ajustado')

# Etiquetas y t√≠tulo
plt.xlabel('iteration')
plt.ylabel('MSE')
plt.title('MSE')
plt.legend()
```

```python
# separate data
n_subgroups = 4
# Inicializar listas para almacenar los subgrupos
x_sub = [[] for _ in range(n_subgroups)]
y_sub = [[] for _ in range(n_subgroups)]
y_pred_sub = [[] for _ in range(n_subgroups)]

for i in range(n_subgroups):
    x_sub[i] = [x[j] for j in range(i, x.size, n_subgroups)]
    y_sub[i] = [y[j] for j in range(i, x.size, n_subgroups)]
    y_pred_sub[i] = [y_pred[j] for j in range(i, x.size, n_subgroups)]

x_sub = [np.array(subgroup) for subgroup in x_sub]
y_sub = [np.array(subgroup) for subgroup in y_sub]
y_pred_sub = [np.array(subgroup) for subgroup in y_pred_sub]
```

```python
# Stochastic gradient Descent
learning_rate = 1e-11
mse_list = []
interations = [i for i in range(10000)]
for iteration in interations:
    mse_list.append(mse(x_sub[iteration % 4],y_sub[iteration % 4] ,y_pred_sub[iteration % 4]))
    y_pred = f(values)
    for i in range(n_subgroups):
        y_pred_sub[i] = np.array([y_pred[j] for j in range(i, x.size, n_subgroups)])
    grad = gradient(y_pred, y, x, h_mean)
    values -= learning_rate * grad

mse_list = np.array(mse_list)
print(f"MSE = {mse(x,y,y_pred)}")
```

```python
plt.figure(figsize=(8, 6))
plt.plot(interations, mse_list, color='red', label='Modelo ajustado')

# Etiquetas y t√≠tulo
plt.xlabel('iteration')
plt.ylabel('MSE')
plt.title('MSE')
plt.legend()
mse_list
```

```python
epochs = 1000
for epoch in range(epochs):
    # Shuffle the data
    indices = np.arange(len(x))
    np.random.shuffle(indices)
    
    for idx in indices:
        x = x[idx]
        y = y[idx]
        x = x - h_mean
        
        # Calcular la predicci√≥n y el gradiente para el punto actual
        y_pred = f(values, x, h_mean)
        grad = gradient(y_pred, y, x, h_mean, values)
        
        # Actualizar los par√°metros
        values -= learning_rate * grad
```
